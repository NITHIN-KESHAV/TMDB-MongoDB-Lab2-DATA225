```python
import numpy as np
import pandas as pd
from ast import literal_eval
import ast

credits_df = pd.read_csv(r"C:\Users\Karthik Shetty\Desktop\Db System for analytics\project\credits.csv")
keywords_df = pd.read_csv(r"C:\Users\Karthik Shetty\Desktop\Db System for analytics\project\keywords.csv")
links_df = pd.read_csv(r"C:\Users\Karthik Shetty\Desktop\Db System for analytics\project\links.csv")
ratings_df = pd.read_csv(r"C:\Users\Karthik Shetty\Desktop\Db System for analytics\project\ratings_small.csv")

movies_data = pd.read_csv(r"C:\Users\Karthik Shetty\Desktop\Db System for analytics\project\movies_metadata.csv", low_memory=False)

# Converting the 'id' column to integers
movies_data["id"] = pd.to_numeric(movies_data['id'], errors='coerce', downcast="integer")

# Processing 'imdb_id' to integer values
movies_data["imdb_id"] = movies_data['imdb_id'].str[2:]
movies_data["imdb_id"] = pd.to_numeric(movies_data['imdb_id'], errors='coerce', downcast="integer")

# Removing entries without an 'id'
movies_data.dropna(subset=["id"], inplace=True)

# Merging additional data
movies_data = movies_data.merge(credits_df, on=["id"], how="left")
movies_data = movies_data.merge(keywords_df, on=["id"], how="left")


```

```python
print("Data after merging credits and keywords:")
display(movies_data.head())
```

```python
# Linking movie IDs between different tables
merged_data = movies_data.merge(links_df, left_on="id", right_on="movieId", how="inner")
print("Data after merging with links on movieId:")
display(merged_data.head())
```

```python
# Checking data shape after linking by imdb_id
shape_data = movies_data.merge(links_df, left_on="imdb_id", right_on="imdbId", how="inner").shape
print("Shape of data after merging on imdb_id:", shape_data)
```

```python
# Cleaning up duplicates and missing values
movies_data.drop_duplicates(inplace=True)
movies_data.dropna(subset=["cast", "crew", "keywords", "popularity"], inplace=True)

# Converting financial figures to numeric and handling zeros
movies_data['budget'] = pd.to_numeric(movies_data['budget'], errors='coerce').replace(0, np.nan)
movies_data['revenue'] = movies_data['revenue'].replace(0, np.nan)

# Refining ratings by averaging them
ratings_df.drop(columns=['userId', 'timestamp'], inplace=True)
avg_ratings = ratings_df.groupby('movieId')['rating'].mean().reset_index()
avg_ratings['rating'] = avg_ratings['rating'].round(1)

print("Average ratings:")
display(avg_ratings.head())

```

```python
movies_data = movies_data.merge(avg_ratings, left_on="id", right_on="movieId", how="left")
movies_data.drop(columns=['movieId'], inplace=True)

# Parsing columns containing JSON
def parse_column(data_column):
    return data_column.fillna('[]').apply(literal_eval).apply(lambda x: [i for i in x] if isinstance(x, list) else [])

movies_data['genres'] = parse_column(movies_data['genres'])
movies_data['production_companies'] = parse_column(movies_data['production_companies'])
movies_data['production_countries'] = parse_column(movies_data['production_countries'])
movies_data['spoken_languages'] = parse_column(movies_data['spoken_languages'])

# Handling cast and crew details
movies_data['cast'] = movies_data['cast'].apply(literal_eval).apply(lambda x: x[:3] if len(x) >= 3 else x)
movies_data['crew'] = movies_data['crew'].apply(literal_eval).apply(lambda x: x[:3] if len(x) >= 3 else x)

print("Sample data for crew:")
display(movies_data['crew'].head())

```

```python
# Exporting the final cleaned dataset
movies_data.to_csv('data.csv', index=False)
```
